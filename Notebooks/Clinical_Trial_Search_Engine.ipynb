```python
# Define the preprocessing function that takes into account medical terms
def preprocess_query_medical(query):
    # Tokenize the query to avoid disrupting medical terms
    tokens = word_tokenize(query)

    # Convert to lowercase
    tokens = [token.lower() for token in tokens]

    # Remove punctuation while preserving numeric data
    tokens = [re.sub(r'[^\w\s]', '', token) for token in tokens]

    # Rejoin tokens into a single string, preserving numeric and medical terms
    preprocessed_query = ' '.join(tokens)

    return preprocessed_query.strip()
```

### **EXTRACTING AND ANALYZING QUERIES FROM XML FILE**

**Overview:**  
This segment of code deals with extracting queries from an XML file. It first sets the path to the file, then uses the `xml.etree.ElementTree` module to parse the XML and extract query information.

- **Setting Up XML File Path:**  
  We define the path to the XML file, pointing to its location in the Colab environment.

- **Parsing XML File:**  
  The `xml.etree.ElementTree.parse` method is used to read the XML file, creating an object (`tree`) that represents the file. The `getroot()` method then retrieves the root of the XML file.

- **Function for Extracting Queries:**  
  The `extract_queries_from_xml` function is defined to iterate over the XML structure, extracting query numbers and their respective text. This data is stored in a dictionary named `queries`, where the key is the query number, and the value is the query text.

- **Executing Query Extraction:**  
  Queries are extracted from the XML root and stored in the `queries` dictionary.

- **Verifying Extraction:**  
  The first three entries from the `queries` dictionary are displayed to check if the extraction process works as intended.

**Purpose:**  
Extracting queries from the XML file is essential for the LLM, and more in general for the search engine, to process and extract the keywords from them later.


```python
# File path for the XML file containing queries
xml_file_path = '/content/queries.xml'

# Parsing the XML file
tree = ET.parse(xml_file_path)  # Parse the XML file
root = tree.getroot()  # Get the root element of the XML file

# Function to extract queries from XML
def extract_queries_from_xml(xml_root):
    queries = {}  # Initialize an empty dictionary for queries
    # Iterate over each 'topic' element in the XML
    for topic in xml_root.iter('topic'):
        topic_number = topic.get('number')  # Get the number attribute of the topic
        # Get the text content of the topic, strip whitespace if text is not None
        topic_text = topic.text.strip() if topic.text is not None else ''
        queries[topic_number] = topic_text  # Add the topic number and text to the dictionary
    return queries

# Extracting and storing queries
queries = extract_queries_from_xml(root)

# Display the first 3 queries to verify extraction
list(queries.items())[:3]
```

### **ENHANCING QUERY PROCESSING FOR LLM THROUGH CONTEXTUAL INFORMATION**

**Overview:**  
We prepare the queries with enriched contextual information for processing by a Large Language Model (LLM). It involves extracting key demographic and health-related data from queries and then reformulating them to include this context, thereby aiding the LLM in more accurately extracting relevant keywords.

- **Code Sections:**
  - **Extraction of Contextual Information:**  
    Functions like `identify_age`, `identify_gender`, and `identify_conditions` are used to extract specific information such as age, gender, and health conditions from the queries. These functions utilize regular expressions and string matching to identify crucial details within the text.
  - **Reformulation of Queries:**  
    The `reformulate_for_llm` function integrates the extracted data into the original queries, creating a structured format that includes contextual tags. This reformulated query is more informative and tailored for the LLM, providing essential context for its keyword extraction task.

**Purpose:**  
The main goal of this process is to optimize the LLM's ability to extract relevant keywords by providing it queries enriched with contextual information. This pre-processing step:
  - Ensures that the LLM has a better understanding of each query's nuances and specific details.
  - Improves the accuracy and relevance of the keyword extraction process.

```python
# Function to identify age from a query
def identify_age(query):
    # Regular expression to find age-related terms
    age_match = re.search(r'\b(\d{1,3})\b(?:\s*|-)*(years|year|months|month|old|year-old)\b', query)
    # Return the age if found, otherwise a default message
    return age_match.group(1) if age_match else "age not specified"

# Function to identify gender from a query
def identify_gender(query):
    # Regular expression to find male-related terms
    if re.search(r'\bmale\b|\bboy\b|\bman\b', query, re.IGNORECASE):
        return 'male'
    # Regular expression to find female-related terms
    elif re.search(r'\bfemale\b|\bgirl\b|\bwoman\b', query, re.IGNORECASE):
        return 'female'
    # Regular expression to find infant-related terms
    elif re.search(r'\binfant\b|\bbaby\b|\bchild\b|\bnewborn\b', query, re.IGNORECASE):
        return 'infant'
    # Default return if no gender is specified
    return "gender not specified"

# Function to identify health conditions from a query
def identify_conditions(query, conditions_dict):
    identified_conditions = []
    # Loop through each condition and its keywords
    for condition, keywords in conditions_dict.items():
        # Check if any keyword is in the query
        if any(keyword in query for keyword in keywords):
            identified_conditions.append(condition)
    # Return identified conditions or a default message
    return identified_conditions if identified_conditions else ["condition not specified"]
```

```python
# Function to reformulate a query for input into a large language model (LLM)
def reformulate_for_llm(query, age, gender, conditions):
    # Constructing the reformulated query with age, gender, and conditions
    reformulated_query = f"Query: {query} | Age: {age} | Gender: {gender} "
    reformulated_query += f"| Conditions: {', '.join(conditions)} "
    return reformulated_query

# Reformulating all queries
reformulated_queries = {}
for key, query in queries.items():
    # Extracting age, gender, and conditions from the query
    age = identify_age(query)
    gender = identify_gender(query)
    conditions = identify_conditions(query, conditions_dict)
    # Reformulating the query using the extracted information
    reformulated_query = reformulate_for_llm(query, age, gender, conditions)
    reformulated_queries[key] = reformulated_query

# Printing the first 5 reformulated queries for inspection
for key, query in list(reformulated_queries.items())[:5]:
    print(f"Key: {key}\nReformulated Query: {query}\n")
```

### **USING GPT-3.5 TURBO FOR KEYWORD EXTRACTION**

**Overview:**  
This code segment sets up and utilizes the OpenAI GPT-3.5 Turbo model to analyze reformulated queries and extract crucial keywords. This process is key to obtaining detailed and relevant information from the queries.

- **API Configuration and Base Prompt:**  
  The OpenAI API key is initialized and a base prompt is defined, guiding the model to analyze medical descriptions in the queries and extract key information such as disease, age, age_type, gender, and health conditions and status.

- **Function for Creating the Prompt:**  
  `create_prompt` combines the base prompt with the reformulated query, preparing it for input into the LLM.

- **Function for Generating Responses:**  
  `generate_response` manages interactions with the GPT-3.5 Turbo model. It includes error handling for rate limits, pausing for 20 seconds before retrying in case of errors.

- **Processing and Saving Responses:**  
  Each reformulated query is processed through the model, and responses are collected and saved in a DataFrame, which is then exported to an Excel file. This file could be used for detailed review and further analysis of the responses.

**Purpose:**  
The aim of this process is to leverage the advanced capabilities of GPT-3.5 Turbo for extracting meaningful keywords from queries. \\
So we transformed unstructured queries into structured, analyzable data.

```python
import time
import openai

# Personal API key for OpenAI - important to keep secure and not share
api_key = ""
openai = openai.OpenAI(api_key=api_key)

# Base prompt for querying the LLM
query_prompt_base = """
Please analyze the following patient's medical description and focus on
extracting key information including Disease, Age, Age_type(years/months/days/…),
Gender, Health Conditions, and Health Status. Include any alternative names,
abbreviations, or synonyms for these terms. If in the remote hyphothesis the
condition is not specified, specify only ‘None’ and not other expressions.
Utilize the additional context provided to enhance the precision of your analysis.
Your response should be a comma-separated list of keywords corresponding to these
categories, which will be crucial for identifying relevant clinical information.
Do not provide elaborate explanations, only the essential keywords derived
from the patient's medical note:
"""

# Function to create the prompt for OpenAI's GPT-3.5 Turbo model
def create_prompt(query_text):
    return [
        {"role": "system", "content": "You are a helpful medical assistant."},
        {"role": "user", "content": query_prompt_base + ' ' + query_text}
    ]

# Function to generate a response from the model
def generate_response(model, query_text):
    if model == "gpt-3.5-turbo":
        try:
            response = openai.chat.completions.create(
                model=model,
                messages=create_prompt(query_text),
                # Parameters for the response
                temperature=0,
                max_tokens=256,
                top_p=1.0,
                frequency_penalty=1.2,
                presence_penalty=1
            )
            answer = response.choices[0].message.content.strip()
            return answer

        # 20-second pause between errors, like the one about rate limit
        except Exception as e:
            print(f'API Error: {e}. Waiting 20 seconds')
            time.sleep(20)
            return generate_response(model, query_text)
    else:
        print('Invalid model selection.')

# Process and print responses for each query
responses = {}
for query_id, query_text in reformulated_queries.items():
    response = generate_response("gpt-3.5-turbo", query_text)
    responses[query_id] = response
    print(f"Response to query {query_id}: {response}")
    time.sleep(5)  # 5-second pause between requests, to limit the error about rate limit

# Convert responses to DataFrame and save to Excel file
import pandas as pd
df_responses = pd.DataFrame.from_dict(responses, orient='index', columns=['Response'])
df_responses.to_excel('responses.xlsx', index=True)
```

### **ANALYSIS AND PROCESSING OF RESPONSES FROM EXCEL FILE**

**Overview:**  
We load responses of the LLM (the structured queries), from an Excel file and we apply different functions to process and analyze the data. Indeed, we extract specific information from the responses to create a dictionary for each query that contains all the values for the keywords we specified as relevant.

- **Excel File Loading:**  
  Responses are loaded from Github into a pandas DataFrame `dfq`, which contains the responses of the LLM to be processed.

- **Function Implementations:**
  - **Extract Age Type:**  
    `extract_age_type` identifies the age type (years, months, days, weeks) from the response using regular expressions, providing clarity on the time frame of the age mentioned.

  - **Determine Health Status:**  
    `is_healthy` assesses if a patient is healthy based on the presence of certain terms indicative of poor health in their health status.

  - **Process Field Values:**  
    `process_field` manages variations of 'None' and 'not specified' in the data fields, ensuring data consistency and cleanliness.

  - **Parse Response into Dictionary:**  
    `parse_response` analyzes each response, extracting crucial information such as age, age_type, gender, health conditions, healthy, diseas and health status, and compiles this into a structured dictionary format for each query.

- **Dictionary Creation for Each Response:**  
  We iterate through each response in the DataFrame, applying `parse_response` to create a dictionary (`queries_dict`) with the velues of the keywords of each query.

- **Result Preparation for Display:**  
  The first five entries are displayed, to inspect the processed data.

**Purpose:**  
The primary aim of this script is to transform the raw response data into a structured, easily analyzable format.


```python
import pandas as pd
import re

# Loading the Excel file 'responses.xlsx' from personal Github
url = 'https://github.com/PatrizioAcquadro/SearchEngineFiles/raw/main/responses.xlsx'
dfq = pd.read_excel(url)

# Function to extract the type of age from the response
def extract_age_type(response):
    match = re.search(r'Age_type:\s*(years|months|days|weeks)', response, re.IGNORECASE)
    return match.group(1).capitalize() if match else None

# Function to determine if a patient is healthy
def is_healthy(health_status):
    unhealthy_terms = ['abnormal', 'low', 'high', 'poor', 'elevated', 'reduced', 'complicated', 'problematic', 'chronic', 'severe']
    return not any(term in health_status.lower() for term in unhealthy_terms)

# Function to process each field, managing 'None' and variations of 'not specified'
def process_field(field_value):
    if field_value:
        lower_value = field_value.lower()
        if any(term in lower_value for term in ['condition not specified', 'none specified', 'not specified', 'none mentioned']):
            return None
    return field_value

# Function to analyze a response and create a dictionary of data
def parse_response(response, query_num):
    age = re.search(r'Age:\s*(\d+)', response)
    gender = re.search(r'Gender:\s*(male|female|infant|both)', response, re.IGNORECASE)
    health_conditions = re.search(r'Health Conditions:\s*(.+)', response)
    health_status = re.search(r'Health Status:\s*(.+)', response)
    disease = re.search(r'Disease:\s*(.+)', response)

    # Constructing a dictionary with extracted data
    query_dict = {
        "conditions": process_field(health_conditions.group(1)) if health_conditions else None,
        "gender": process_field(gender.group(1)) if gender else None,
        "healthy": is_healthy(health_status.group(1)) if health_status else None,
        "age_number": int(age.group(1)) if age else None,
        "age_type": extract_age_type(response),  # Using the new function extract_age_type
        "disease": process_field(disease.group(1)) if disease else None,
        "health_status": process_field(health_status.group(1)) if health_status else None
    }

    return query_dict

# Creating a dictionary for each response in the DataFrame
queries_dict = {}
for index, row in dfq.iterrows():
    response = row['Response']
    queries_dict[index] = parse_response(response, index)

# Preparing results for display
first_three_queries_results = {k: queries_dict[k] for k in sorted(queries_dict)[:5]}
first_three_queries_results
```

```python
def get_gender_age_clinical_trials(query):
  gender = query[gender]
  age_value = query[age_value]
  age_type = query[age_type]
  if not type(age_value) is int and gender is not None:
    raise RuntimeError('age_value must be an integer or None')
  if age_value < 0:
    raise RuntimeError('age_value must be positive or equal to 0')
  if (age_type is not None and age_value is None):
    raise RuntimeError('age_value must not be None if age_type is not None')

  query = """
    SELECT DISTINCT ?clinicalTrial
    WHERE {
  """

  # == Gender ==
  if gender == "male":
    query += " ?clinicalTrial a ct:MaleClinicalTrial . "
  elif gender == "female":
    query += " ?clinicalTrial a ct:FemaleClinicalTrial . "
  elif gender == "all":
    query += " ?clinicalTrial a ct:MaleClinicalTrial . "
    query += " ?clinicalTrial a ct:FemaleClinicalTrial . "
  elif gender is not None:
    raise RuntimeError('gender must be either "male", "female", "all" or None')

  # == Age ==
  if age_type == "years":
    query += " ?clinicalTrial a ct:YearClinicalTrial . "
    query += " ?clinicalTrial ct:eligibilityMaximumYears ?maximumAge . "
    query += " ?clinicalTrial ct:eligibilityMinimumYears ?minimumAge . "
    query += f" FILTER(?minimumAge <= {age_value} && ?maximumAge >= {age_value}) . "
  elif age_type == "months":
    query += " ?clinicalTrial a ct:MonthClinicalTrial . "
    query += " ?clinicalTrial ct:eligibilityMaximumMonths ?maximumAge . "
    query += " ?clinicalTrial ct:eligibilityMinimumMonths ?minimumAge . "
    query += f" FILTER(?minimumAge <= {age_value} && ?maximumAge >= {age_value}) . "
  elif age_type == "weeks":
    query += " ?clinicalTrial a ct:WeekClinicalTrial . "
    query += " ?clinicalTrial ct:eligibilityMaximumWeeks ?maximumAge . "
    query += " ?clinicalTrial ct:eligibilityMinimumWeeks ?minimumAge . "
    query += f" FILTER(?minimumAge <= {age_value} && ?maximumAge >= {age_value}) . "
  elif age_type == "days":
    query += " ?clinicalTrial a ct:DayClinicalTrial . "
    query += " ?clinicalTrial ct:eligibilityMaximumDays ?maximumAge . "
    query += " ?clinicalTrial ct:eligibilityMinimumDays ?minimumAge . "
    query += f" FILTER(?minimumAge <= {age_value} && ?maximumAge >= {age_value}) . "
  elif age_type == "hours":
    query += " ?clinicalTrial a ct:HourClinicalTrial . "
    query += " ?clinicalTrial ct:eligibilityMaximumHours ?maximumAge . "
    query += " ?clinicalTrial ct:eligibilityMinimumHours ?minimumAge . "
    query += f" FILTER(?minimumAge <= {age_value} && ?maximumAge >= {age_value}) . "
  elif age_type == "minutes":
    query += " ?clinicalTrial a ct:MinuteClinicalTrial . "
    query += " ?clinicalTrial ct:eligibilityMaximumMinutes ?maximumAge . "
    query += " ?clinicalTrial ct:eligibilityMinimumMinutes ?minimumAge . "
    query += f" FILTER(?minimumAge <= {age_value} && ?maximumAge >= {age_value}) . "
  elif age_type is not None:
    raise RuntimeError('age_type must be either "years", "months", "weeks", "days", "hours", "minutes" or None')

  query += "}"
  return get_clinical_trial_ids(graph.query(query))
```

Now that we obtained the IDs of the right trials for age and sex, we can check the exclusion and inclusion criteria. In particular, these criteria are referred to the 'condition' field of the dictionary, which includes conditions extracted with the LLM, which can be both in the inclusion or exclusion criteria. Also in this case there is a transformer (which will be part of the final retrieval pipeline) that query the KG. This query is making sure that the list of criteria (the condition of the query dictionary) matches with the inclusion criteria AND doesn't match with the exclusion ones.

```python
def get_criteria_clinical_trials(query):
  if query[conditions] is None:
    return
  criteria = list(query[conditions].split(", "))
  if not type(criteria) is list:
    raise RuntimeError('criteria must be a list')

  query = """
    SELECT DISTINCT ?clinicalTrial
    WHERE {
  """

  # == Inclusion criteria ==
  query += " ?clinicalTrial a ct:ClinicalTrial . "
  query += " { "
  for i in range(len(criteria)):
    if i == len(criteria) - 1:
      query += f" ?clinicalTrial ct:eligibilityInclude ct:{criteria[i]} . "
      continue

    query += f" ?clinicalTrial ct:eligibilityInclude ct:{criteria[i]} . "  + " } UNION { "
  query += " } "

  # == Exclusion criteria ==
  query += " FILTER NOT EXISTS { { "
  for i in range(len(criteria)):
    if i == len(criteria) - 1:
      query += f" ?clinicalTrial ct:eligibilityExclude ct:{criteria[i]} . "
      continue

    query += f" ?clinicalTrial ct:eligibilityExclude ct:{criteria[i]} . "  + " } UNION { "
  query += " } } "

  query += "}"
  return get_clinical_trial_ids(graph.query(query))
```
